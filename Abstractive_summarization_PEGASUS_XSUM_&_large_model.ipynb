```python
##Import BeautifulSoup libraries
from bs4 import BeautifulSoup
import requests
##Import Pegasus models and tokenizer
from transformers import PegasusForConditionalGeneration,PegasusTokenizer
import gradio as gr
```

##Scrape Scientific Article 

```python
url='https://towardsdatascience.com/my-absolute-go-to-for-sentiment-analysis-textblob-3ac3a11d524'
response=requests.get(url)
```

```python
#Perse content
soup=BeautifulSoup(response.text, 'html')
print(soup.prettify())
```

```python
##Extract required Titles and text
results=soup.find_all(['h1','h2','p','b'])
results
```

```python
## Remove all the label from the text and joined back text
text=[result.text for result in results]
Article=' '.join(text)
Article
```

## Perforrm Abstractive Summarization using Pegasus-xsum

```python
#Create tokenizer
tokenizer = PegasusTokenizer.from_pretrained('google/pegasus-xsum')
```

```python
#Load model
model=PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
```

```python
##Create Tokens - number representation of our text
tokens=tokenizer(Article,truncation=True,padding="longest",return_tensors="pt")
```

```python
tokens
```

```python
#Summarize
summary= model.generate(**tokens)
```

```python
{**tokens}
```

```python
#Decode summary
tokenizer.decode(summary[0])
```

## Text generator and Interface

```python
# Function for text generation
def generate_xsum_text(prompt):
    input_id = tokenizer(prompt,truncation=True,padding="longest",return_tensors="pt")
    output_text = model.generate(**input_id)
    P_xsum_summary = tokenizer.decode(output_text[0], skip_special_tokens=True)
    return P_xsum_summary
```

```python
# Gradio interface
input_text = gr.inputs.Textbox(lines=3, label="Input Text")
New_text = gr.outputs.Textbox(label="Generated Text")

iface = gr.Interface(
    fn=generate_xsum_text,
    inputs=input_text,
    outputs=New_text,
    title="Pegasus_xsum Text Summarization",
    description="Abstractive text summarization using Pegasus_xsum model.",
)

iface.launch()
```
