```python
## Import Libraries
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
```

```python
#Load summarizationn pipeline
summarizer=pipeline('summarization')
```

```python
URL="https://www.edureka.co/blog/what-is-a-neural-network/#Difference%20Between%20AI,%20ML%20and,%20DL"
```

```python
## Request for the website
source=requests.get(URL).text
```

```python
##BeautifulSoup -For parsing webpage
soup=BeautifulSoup(source,'html')
#print(soup.prettify())
```

```python
results=soup.find_all(['h1','h2','p'])
#results
```

```python
text=[result.text for result in results]
Article=' '.join(text)
#Article
```

```python
##Chunk Text
Article =Article .replace('.','.<eos>')
Article =Article .replace('!','!<eos>')
Article =Article .replace('?','?<eos>')
sentences=Article.split('<eos>')
```

```python
sentences
```

```python
max_chunk=500
current_chunk=0
chunks=[]

for sentence in sentences:
  if len(chunks)==current_chunk +1:
    if len(chunks[current_chunk])+len(sentence.split(' '))<=max_chunk:
      chunks[current_chunk].extend(sentence.split(' '))
    else:
        current_chunk += 1
        chunks.append(sentence.split(' '))
  else:
          print(current_chunk)
          chunks.append(sentence.split(' '))
```

```python
for chunk_id in range(len(chunks)):
  chunks[chunk_id]=' '.join(chunks[chunk_id])
```

```python
chunks[1]
```

```python
len(chunks[1].split(' '))
```

```python
##Summmarize chunks
res=summarizer(chunks,max_length=120,min_length=30,do_sample=False)
```

```python
#Output to Text file
Text = ' '.join([summ['summary_text'] for summ in res])
Text
```
